```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
visa_df=pd.read_csv(r"C:\Users\HP\EDA\dataset\Visadataset.csv")
visa_df
```

```python
min(visa_df['prevailing_wage'])
max(visa_df['prevailing_wage'])
```

```python
index=["count","min","max","mean","median","std","25p","50p","75 p"]
num_col=visa_df.select_dtypes(exclude="object").columns
num_df=pd.DataFrame()

for i in num_col:
    count=len(visa_df[i])
    mini=min(visa_df[i])
    maxi=max(visa_df[i])
    mean=round(visa_df[i].mean(),2)
    median=round(visa_df[i].median(),2)
    std=round(visa_df[i].std(),2)
    per_25=round(np.percentile(visa_df[i],25),2)
    per_50=round(np.percentile(visa_df[i],50),2)
    per_75=round(np.percentile(visa_df[i],75),2)
    l=[count,mini, maxi,mean,median,std,per_25,per_50,per_75]
    num=pd.DataFrame(l,index=index,columns=[i])
    num_df=pd.concat([num,num_df],axis=1)
num_df
```
